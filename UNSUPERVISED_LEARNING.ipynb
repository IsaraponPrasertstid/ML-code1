```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
numeric_columns = [
    'TotalReplies', 'TotalResponses', 'TotalPosts', 'AverageWordsPerPost', 'LikeRate', 
    'QuestionsAsked', 'URLSharing', 'MeanPostsPerThread', 'ConversationInitiationRate', 
    'PostingDiversity', 'BidirectionalNetwork'
]
social_standardize = social[:]
social_standardize[numeric_columns] = scalar.fit_transform(social_standardize[numeric_columns])
social_standardize
```

```python
from sklearn.cluster import KMeans

k = 10
inertias = []
for k in range(1,k+1):
    kmm = KMeans(n_clusters=k, random_state=15)
    kmm.fit(social_standardize)
    inertias.append(kmm.inertia_)


plt.figure()
plt.plot(range(1, k+1), inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The elbow method showing the optimal k')
```

```python
from sklearn.metrics import silhouette_score

# compare between 3 can 4
kmeans_model = KMeans(n_clusters=4, random_state=15)
kmeans_model.fit(social_standardize)
cluster_labels = kmeans_model.predict(social_standardize)

print(f"Kmean Silhouette Scores: {silhouette_score(social_standardize, cluster_labels)}")
```

```python
from sklearn.metrics import silhouette_score

# compare between 3 can 4
kmeans_model = KMeans(n_clusters=3, random_state=15)
kmeans_model.fit(social_standardize)
cluster_labels = kmeans_model.predict(social_standardize)

print(f"Kmean Silhouette Scores: {silhouette_score(social_standardize, cluster_labels)}")
```

The Silhouette Scores when k = 3 is higher than 4, so i will use k=3 to model the data

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ลดมิติข้อมูลเป็น 3 มิติ (PCA)
pca = PCA(n_components=3)
X_pca = pca.fit_transform(social_standardize)

# สร้างกราฟ 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# แบ่งกลุ่มข้อมูลที่ได้จาก KMeans โดยใช้สีที่ต่างกัน
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=cluster_labels, cmap='viridis')

# ใส่ชื่อแกน
ax.set_xlabel('PCA 1')
ax.set_ylabel('PCA 2')
ax.set_zlabel('PCA 3')

# ใส่ colorbar เพื่อแสดง cluster label
fig.colorbar(scatter, ax=ax)

# แสดงกราฟ
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import GridSearchCV
import numpy as np

# กำหนดพารามิเตอร์ที่ต้องการ tune
param_grid = {
    'n_components': [2, 3, 4, 5, 6],
    'covariance_type': ['full', 'tied', 'diag', 'spherical'],
    'max_iter': [100, 200, 300],
    'tol': [1e-3, 1e-4, 1e-5],
}

# สร้าง GMM model
gmm = GaussianMixture(random_state=15)

# ใช้ GridSearchCV เพื่อค้นหาค่าพารามิเตอร์ที่ดีที่สุด
grid_search = GridSearchCV(gmm, param_grid, cv=3, verbose=1) # ในกรณีนี้ neg_log_loss คือ log-likelihood ที่ใช้ในการเลือกโมเดลที่ดีที่สุด

grid_search.fit(social_standardize)

# แสดงค่าพารามิเตอร์ที่ดีที่สุด
print("Best Parameters: ", grid_search.best_params_)

# นำโมเดลที่ดีที่สุดไปใช้
best_gmm = grid_search.best_estimator_

# ใช้ GMM ที่ดีที่สุดมาทำนายและทำการ plot
gmm_labels = best_gmm.predict(social_standardize)

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

# ลดมิติข้อมูลให้เป็น 3 มิติ
pca = PCA(n_components=3)
X_pca = pca.fit_transform(social_standardize)

# Plot GMM Clusters ใน 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# สร้าง scatter plot สำหรับ GMM clusters
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=gmm_labels, cmap='viridis')

# เพิ่ม title และ labels
ax.set_title('Tuned Gaussian Mixture Model Clustering (3D)')
ax.set_xlabel('PCA 1')
ax.set_ylabel('PCA 2')
ax.set_zlabel('PCA 3')

# เพิ่ม color bar
fig.colorbar(scatter)

# แสดงกราฟ
plt.show()

```

## T-test to compare performance between baseline model (Kmean clustering) and GMM

```python
from sklearn.model_selection import KFold
from sklearn.metrics import silhouette_score
import numpy as np

# สมมติว่า social_standardize เป็น numpy array หรือ DataFrame ที่ไม่มีปัญหา
X = social_standardize.values  # แปลงเป็น numpy array เพื่อใช้ในการ split กับ KFold

cv = KFold(n_splits=10, random_state=15, shuffle=True) 
scoresbl = []  # สำหรับเก็บ silhouette score ของ KMeans
scoresgmm = []  # สำหรับเก็บ silhouette score ของ GMM

# ใช้ KFold เพื่อแบ่งข้อมูล
for train_index, test_index in cv.split(X):
    Xtr, Xtest = X[train_index], X[test_index]  # ใช้ข้อมูลฝึก (Xtr) และข้อมูลทดสอบ (Xtest)
    
    # Train GMM กับข้อมูลฝึก
    modelgmm = best_gmm.fit(Xtr)
    
    # Train KMeans กับข้อมูลฝึก
    modelbl = kmeans_model.fit(Xtr)
    
    # ทำนาย cluster labels
    cluster_lb_gmm = modelgmm.predict(Xtest)  # ใช้ข้อมูลทดสอบ
    cluster_label_bl = modelbl.predict(Xtest)  # ใช้ข้อมูลทดสอบ
    
    # คำนวณ silhouette score สำหรับ GMM และ KMeans
    gmm_silhouette = silhouette_score(Xtest, cluster_lb_gmm)  # สำหรับ GMM
    silhouette_baseline = silhouette_score(Xtest, cluster_label_bl)  # สำหรับ KMeans
    
    # เก็บผลลัพธ์ silhouette score
    scoresgmm.append(gmm_silhouette)
    scoresbl.append(silhouette_baseline)
```

```python
from scipy import stats
# t-test
tstat_bl_gmm, pval_bl_gmm = stats.ttest_rel(scoresbl, scoresgmm)

print(f"GMM vs Kmeans: t-statistic={tstat_bl_gmm.mean():.2f}, p-value={pval_bl_gmm.mean():.3f}")

if pval_bl_gmm.mean() < 0.05:
    print("Outcome of the paired t-test shows that the performance of the models are significantly different.")
else:
    print("Outcome of the paired t-test shows that I cannot reject the hypothesis. The performance of the models is not significantly different.")
```
