```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow.compat.v1 as tf
import requests
tf.disable_v2_behavior()
sess = tf.Session()
```

```python
housing_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
housing_header = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX',
  'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
cols_used = ['CRIM', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS',
'TAX', 'PTRATIO', 'B', 'LSTAT']
num_features = len(cols_used)
```

```python
line = '0.00632  18.00  2.310   0.0  0.538'
parts = line.split(' ')
print(parts)
```

```python
housing_file = requests.get(housing_url)
lines = housing_file.text.strip().split('\n')  # ตัด \n ท้ายสุดออกด้วย
housing_data = []

for line in lines:
    if len(line) < 1:
        continue  # ข้ามบรรทัดว่าง
    parts = line.split(' ')
    clean_parts = [float(x) for x in parts if len(x) >= 1]
    housing_data.append(clean_parts)

```

x_vals: ค่าคุณลักษณะทั้งหมดที่เลือกใช้ (ตาม cols_used)

```python
y_vals = np.transpose([np.array([y[13] for y in housing_data])])
# np.transpose() ใช้เพื่อเปลี่ยนลิสต์จากแนวนอนเป็นแนวตั้ง (เป็นแถวเดียว) เพื่อให้เหมาะกับการใช้งานกับ TensorFlow หรือโมเดลต่าง ๆ ที่ต้องการข้อมูลเป็น แนวตั้ง (1D array)
# NumPy array ซึ่งเหมือนกับลิสต์ แต่สามารถทำคำนวณแบบเวกเตอร์ได้ง่าย เช่น การบวก ลบ คูณ หารทีละหลายตัวพร้อมกัน

x_vals = np.array([[x for i,x in enumerate(y) if housing_header[i] in cols_used] for y in housing_data]) 
# สำหรับแต่ละบ้าน (y) จะสร้างลิสต์ของค่าของ x (ค่าจากคอลัมน์ต่าง ๆ) ที่ตรงกับคอลัมน์ใน cols_used
```

```python
# Normalize ข้อมูลให้มีค่าอยู่ระหว่าง 0-1 โดยใช้ min-max scaling
# ปรับข้อมูลให้อยู่ในช่วง 0 ถึง 1 (Normalization) แบบ min-max scaling ซึ่งเป็นขั้นตอนสำคัญมากใน Machine Learning โดยเฉพาะกับอัลกอริธึมที่อ่อนไหวต่อค่าขนาด เช่น KNN หรือ Neural Network
# a.min(0) → หาค่าต่ำสุด "ในแต่ละคอลัมน์":


# [1, 2, 3]   ← แถวที่ 1  
# [4, 5, 6]   ← แถวที่ 2  
# ↓ ↓ ↓
# [1, 2, 3]   ← ค่า min ในแต่ละคอลัมน์

x_vals = (x_vals - x_vals.min(0)) / x_vals.ptp(0)

# เอา ค่ามากสุด (peak) ลบด้วย ค่าต่ำสุด (trough)
# ทำแบบนี้ ในแต่ละคอลัมน์ (เพราะ axis=0)
# ใช้เพื่อหา ช่วงของค่าฟีเจอร์ (range) → สำหรับการ Normalize ให้เป็น 0-1


# x = np.array([
#     [10, 20, 30],
#     [40, 50, 60],
#     [70, 80, 90]
# ])

# การทำงานของ x.ptp(0):
# คอลัมน์ที่ 1: 70 - 10 = 60
# คอลัมน์ที่ 2: 80 - 20 = 60
# คอลัมน์ที่ 3: 90 - 30 = 60

# สูตรการทำ Min-Max Normalization
# จะปรับให้ข้อมูลแต่ละคอลัมน์อยู่ในช่วง [0, 1]
# ทำให้ฟีเจอร์แต่ละตัวมีน้ำหนักเท่า ๆ กัน ไม่ให้ฟีเจอร์ที่ค่ามากกว่าครอบงำฟีเจอร์อื่น
```

```python
# แบ่งข้อมูลออกเป็นชุดฝึก (Train set) และ ชุดทดสอบ (Test set) 

# เลือก index แบบสุ่ม จำนวน 80% ของข้อมูลทั้งหมด (len(x_vals)*0.8)
# replace=False หมายถึง ไม่สุ่มซ้ำ
# ได้ผลลัพธ์เป็น index ที่ใช้ฝึกโมเดล

# np.random.choice(array, size, replace=False)
# พารามิเตอร์	  ความหมาย
# array	        แหล่งข้อมูลที่ต้องการสุ่มจาก (สามารถเป็นจำนวนเต็มก็ได้ เช่น 10 คือ [0,1,...,9])
# size	        จำนวนตัวอย่างที่ต้องการสุ่ม
# replace	    ถ้า True → สุ่มซ้ำได้, ถ้า False → สุ่มแบบไม่ซ้ำ

train_indices = np.random.choice(len(x_vals), round(len(x_vals)*0.8), replace=False)


# หาค่าที่ไม่ได้อยู่ใน train_indices
# 1. range(len(x_vals))
# สร้างลิสต์ของ index ทั้งหมด
# เช่น ถ้า x_vals มี 5 แถว → ได้ [0, 1, 2, 3, 4]
# 2. set(...)
# แปลงเป็น set เพื่อให้ลบกันได้ง่าย เช่น {0,1,2,3,4}
# 3. set(...) - set(train_indices)
# เอา index ทั้งหมด ลบกับ index ที่ใช้ฝึก
# เหลือเฉพาะ index ที่ยังไม่ได้ใช้ → คือชุดทดสอบ
# 4. list(...) และ np.array(...)
# แปลงกลับเป็น list → แล้วแปลงเป็น NumPy array → เพื่อใช้กับ x_vals[...]
test_indices = np.array(list(set(range(len(x_vals))) - set(train_indices)))


# นำ index ที่ได้มาเลือกเฉพาะแถวของข้อมูล x_vals และ y_vals
x_vals_train = x_vals[train_indices]
x_vals_test = x_vals[test_indices]
y_vals_train = y_vals[train_indices]
y_vals_test = y_vals[test_indices]
```

```python
# ตั้งค่า TensorFlow
# กำหนดค่า k = 4 สำหรับ k-NN
# สร้าง placeholder สำหรับข้อมูลฝึก, ข้อมูลทดสอบ, ค่าที่ต้องการทำนาย
k = 4

# ตั้งค่า batch size = ขนาดของชุดทดสอบ (จะทำนายทั้งหมดในคราวเดียว)
batch_size = len(x_vals_test)

# placeholder
# "เตรียมช่องรับข้อมูล 2 คอลัมน์
# และจำนวนแถวเท่าไหร่ก็ได้ (None)
# แล้วฉันจะใส่ข้อมูลลงในช่องนี้ภายหลัง"

# ทำไมต้องใช้ placeholder?
# เพราะใน TensorFlow v1 ต้องสร้าง กราฟการคำนวณล่วงหน้า ก่อน
# เราจะ "ใส่ข้อมูลจริง" ตอน sess.run(...)
x_data_train = tf.placeholder(shape=[None, num_features], dtype=tf.float32)
x_data_test = tf.placeholder(shape=[None, num_features], dtype=tf.float32)
y_target_train = tf.placeholder(shape=[None, 1], dtype=tf.float32)
y_target_test = tf.placeholder(shape=[None, 1], dtype=tf.float32)
```

```python
# คำนวณระยะห่างและเลือก k ใกล้สุด
# expand_dims เพื่อเพิ่มมิติจะทำให้ x_data_test กลายเป็นเทนเซอร์ 3D ที่มีขนาด (n_samples_test, 1, num_features) 
# การใช้ expand_dims จะเพิ่มมิติที่ตำแหน่งที่ 1 ของ x_data_test ซึ่งทำให้ข้อมูลเปลี่ยนจากเทนเซอร์ 2D (ตัวอย่าง x ฟีเจอร์) ให้กลายเป็นเทนเซอร์ 3D ที่มีขนาด (n_samples_test, 1, num_features)
# พื่อให้สามารถทำการคำนวณระยะห่างระหว่าง x_data_train และ x_data_test ได้


# เราต้องการคำนวณระยะห่าง (distance) ระหว่างตัวอย่าง ทุกตัวใน x_data_train กับทุกตัวใน x_data_test.
# หากข้อมูลของเราเป็น 2D tensor (x_data_train และ x_data_test), การลบระหว่าง 2D tensor โดยตรงจะไม่ได้ผลเนื่องจากจำนวนมิติของมันไม่ตรงกัน (จำนวนตัวอย่างใน x_data_train และ x_data_test อาจแตกต่างกัน).


# x_data_train = [[1, 2, 3], [4, 5, 6]]
# x_data_test = [[2, 2, 2], [3, 3, 3]]
# x_data_test_expanded = [[[2, 2, 2], [2, 2, 2]],
#                         [[3, 3, 3], [3, 3, 3]]]

# manhatt
# tf.subtract(x_data_train, x_data_test_expanded) = [
#    [[1-2, 2-2, 3-2], [1-2, 2-2, 3-2]],
#    [[4-3, 5-3, 6-3], [4-3, 5-3, 6-3]]
# ]
# ได้
# [[[ -1,  0,  1], [-1,  0,  1]],
#  [[  1,  2,  3], [ 1,  2,  3]]]

# tf.abs
# [
#   [[1, 0, 1], [2, 3, 4]],
#   [[2, 1, 0], [1, 2, 3]]
# ]

# reduce_sum หาผลรวมของค่าต่าง ๆ ตามแกน (axis) ที่กำหนดไว้ใน Tensor หรืออาร์เรย์หลายมิติ เป็นผลรวมของระยะห่าง:
# distance = [
#   [2, 2],
#   [6, 6]
# ] 


distance = tf.reduce_sum(tf.abs(tf.subtract(x_data_train, tf.expand_dims(x_data_test,1))), reduction_indices=2)

# คำนวณ Euclidean Distance
# distance_euclidean = tf.sqrt(tf.reduce_sum(tf.square(tf.subtract(x_data_train, tf.expand_dims(x_data_test, 1))), axis=2))


# เลือก k ตัวอย่างในชุดฝึก (train) ที่มีระยะทาง ใกล้ที่สุด กับแต่ละตัวอย่างในชุดทดสอบ (test) โดยใช้ TensorFlow ฟังก์ชัน tf.nn.top_k
# ❷ tf.negative(distance)
# เปลี่ยนค่าระยะห่างทั้งหมดให้เป็นค่าติดลบ
# เพราะ tf.nn.top_k จะเลือก ค่าสูงสุด (เราอยากได้ "ใกล้ที่สุด" → คือ "ค่าต่ำสุด" → จึงต้องใส่ลบ)
top_k_xvals, top_k_indices = tf.nn.top_k(tf.negative(distance), k=k)
```

```python
top_k_xvals
```

```python
# ในขั้นตอนถัดไปคุณจะนำค่าระยะห่างเหล่านี้ไปใช้ในการเลือก k เพื่อนบ้านที่ใกล้ที่สุด (top-k nearest neighbors) และคำนวณค่าเฉลี่ยแบบถ่วงน้ำหนักเพื่อทำการทำนายผล.
# 1. คำนวณผลรวมของ top_k_xvals ตามมิติ 1 (ที่เราต้องการหาผลรวมของระยะห่างในแต่ละแถว)
x_sums = tf.expand_dims(tf.reduce_sum(top_k_xvals, 1), 1)  # เพิ่มมิติที่ 1 (ทำให้เป็น 2D tensor)

# 2. ทำการสร้างเทนเซอร์ที่มีค่าทั้งหมดเป็น 1 เพื่อให้สามารถคูณกับ x_sums (repeated sum)

# top_k_xvals = [
#   [-2.0, -2.0],
#   [-6.0, -6.0]
# ]

# x_sums = [
#   [-4.0],
#   [-12.0]
# ]

# x_val_weights = top_k_xvals / x_sums

# TensorFlow จะ error เพราะ shape ไม่ตรง
# เราจึง สร้าง x_sums_repeated ให้มีขนาดเท่ากันกับ top_k_xvals:

# tf.ones([1, 2]) = [[1.0, 1.0]]

    
# x_sums_repeated = [
#   [-4.0, -4.0],
#   [-12.0, -12.0]
# ]


# tf.matmul คือฟังก์ชันของ TensorFlow สำหรับการ คูณเมทริกซ์ (matrix multiplication) โดยเฉพาะครับ
# ไม่ใช่การคูณแบบ element-wise (คูณตามตำแหน่ง) แต่เป็น คูณแบบเมทริกซ์จริง ๆ ตามหลักคณิตศาสตร์เชิงเส้น


x_sums_repeated = tf.matmul(x_sums, tf.ones([1, k], tf.float32))  # ขยาย x_sums ให้มีขนาดเท่ากับ k

# 3. คำนวณค่าน้ำหนัก (weights) ของแต่ละเพื่อนบ้านจากระยะห่าง
# ค่าน้ำหนักจะเป็นการหารค่า top_k_xvals ด้วยผลรวมที่เราคำนวณจากทุก ๆ เพื่อนบ้าน
#  เพิ่มมิติใหม่ที่ตำแหน่งที่ 1 ทำให้ tensor เปลี่ยนจาก [batch_size, k] → [batch_size, 1, k] 
# การ เพิ่มมิติใหม่ที่ตำแหน่งที่ 1 (tf.expand_dims(..., 1)) เป็นการแปลงรูป tensor จาก [batch_size, k] เป็น [batch_size, 1, k] เพื่อให้สามารถนำไปใช้ในการคูณกับค่าจริง (y_target) ในขั้นตอนถัดไปได้
# เพราะในขั้นตอนถัดไป  จะต้องเอา tensor ขนาด [batch_size, 1, k]
# ไปคูณกับ [batch_size, k, 1] เพื่อให้ได้ค่าเฉลี่ยแบบถ่วงน้ำหนักที่เป็น [batch_size, 1, 1] ->สำหรับแต่ละตัวอย่างใน batch (ที่มีขนาดเป็น batch_size), จะได้ผลลัพธ์เป็น หนึ่งค่าทำนาย (single predicted value).

x_val_weights = tf.expand_dims(tf.divide(top_k_xvals, x_sums_repeated), 1)  # ค่าน้ำหนัก


# 4. ดึงค่าของ y_target_train ที่ตรงกับ top_k_indices ซึ่งคือค่าจริงของเพื่อนบ้านที่ใกล้ที่สุด
# top_k_indices คือดัชนี (indices) ของ k เพื่อนบ้านที่ใกล้ที่สุด จากชุดฝึก (training set) สำหรับตัวอย่างในชุดทดสอบ (test set)
# y_target_train คือค่าจริง (true labels) ของตัวอย่างในชุดฝึก
# tf.gather ใช้เพื่อดึงค่าจริงที่ตรงกับดัชนีใน top_k_indices สำหรับแต่ละตัวอย่างในชุดทดสอบ
top_k_yvals = tf.gather(y_target_train, top_k_indices)  # ค่าจริงจากเพื่อนบ้านที่ใกล้ที่สุด

# 5. คำนวณค่าทำนาย (prediction) โดยการคูณน้ำหนักกับค่าจริง
# x_val_weights: ค่าน้ำหนัก (weights) ที่คำนวณจากระยะห่างระหว่างตัวอย่างในชุดฝึกและชุดทดสอบ โดยจะมีขนาดเป็น [batch_size, 1, k]
# top_k_yvals: ค่าจริงของเพื่อนบ้านที่ใกล้ที่สุดที่ได้จากขั้นตอนที่แล้ว โดยมีขนาดเป็น [batch_size, k, 1]
# tf.matmul(x_val_weights, top_k_yvals): การคูณเมทริกซ์ระหว่าง x_val_weights และ top_k_yvals จะให้ผลลัพธ์เป็น [batch_size, 1, 1] ซึ่งเป็นการคำนวณ ค่าเฉลี่ยแบบถ่วงน้ำหนัก สำหรับแต่ละตัวอย่างในชุดทดสอบ
# tf.squeeze(..., axis=1): หลังจากที่ได้ผลลัพธ์เป็น [batch_size, 1, 1], การใช้ tf.squeeze() เพื่อลบมิติที่ไม่จำเป็น (มิติที่มีขนาดเป็น 1) ออก จะทำให้ผลลัพธ์สุดท้ายเป็น [batch_size] ซึ่งเป็น ค่าทำนายสุดท้าย (predictions) สำหรับแต่ละตัวอย่าง
prediction = tf.squeeze(tf.matmul(x_val_weights, top_k_yvals), axis=1)  # คำนวณค่าทำนายสุดท้าย

```

```python
# คำนวณ Mean Squared Error (MSE)

# tf.reduce_sum(...):
# คำสั่งนี้ใช้ในการ หาผลรวม ของค่าผลต่างยกกำลังสอง (error squared) ที่ได้จากขั้นตอนที่แล้ว:
    
# tf.divide(..., batch_size):
# ขั้นตอนนี้ใช้ในการ หารผลรวมของ squared errors ด้วยจำนวน batch_size เพื่อคำนวณ ค่าเฉลี่ย (mean) ของค่า squared errors สำหรับแต่ละตัวอย่างในชุดทดสอ

mse = tf.divide(tf.reduce_sum(tf.square(tf.subtract(prediction, y_target_test))), batch_size)
```
